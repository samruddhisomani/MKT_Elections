```python
yearmask=anes['VCF0004'].isin([2008,2012])
```

```python
anes_year=anes[yearmask]
```

```python
anes_final=anes_year.dropna(axis=1,how='any')
```

```python
left=['VCF0901a','VCF0004']
right=['GEO.id2','Year']
anes_acs=pd.merge(anes_final,acs,left_on=left, right_on=right,how='outer')
```

```python
adspend=pd.read_csv('adspend_all_years_edited2.csv')
```

```python
ad=['FIPS','Year']
merge_adspend=pd.merge(anes_acs,adspend,left_on=right,right_on=ad,how='outer')
```

```python
votes=pd.read_csv('Votes_all_years.csv')
```

```python
votes['FIPS_votes'] = votes["STATE"].map(lambda x: us.states.mapping('name','fips')[x])
```

```python
votes['FIPS_votes']=votes['FIPS_votes'].astype(int)
```

```python
vote=['FIPS_votes','YEAR']
merge_votes=pd.merge(merge_adspend,votes,left_on=ad,right_on=vote,how='outer')
```

```python
turnout=pd.read_csv('turnout_all_years.csv')
```

```python
turnout['FIPS_turnout']=turnout['State'].map(lambda x: us.states.mapping('name','fips')[x])
```

```python
turnout['FIPS_turnout']=turnout['FIPS_turnout'].astype(int)
turnout_join=['FIPS_turnout','Year']
merge_turnout=pd.merge(merge_votes, turnout,left_on=vote, right_on=turnout_join)
```

```python
econ=pd.read_csv('econAgg.csv')
```

```python
econ['FIPS_econ']=econ['Fips'].map(lambda x: x/1000)
```

```python
econ1=econ.drop(['Fips','Personal'],1)
```

```python
econ1.rename(columns={'level_1':'year'}, inplace=True)
```

```python
econ_join=['FIPS_econ','year']
merge_econ=pd.merge(merge_turnout,econ1,left_on=vote,right_on=econ_join)
```
